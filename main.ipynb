- **track_id**: O ID do Spotify para a faixa  
- **artists**: Os nomes dos artistas que realizaram a faixa. Se houver mais de um artista, eles são separados por ponto e vírgula (;)  
- **album_name**: O nome do álbum em que a faixa aparece  
- **track_name**: Nome da faixa  
- **popularity**: A popularidade de uma faixa é um valor entre 0 e 100, sendo 100 o mais popular. É calculada por algoritmo e se baseia principalmente no número total de reproduções que a faixa teve e em quão recentes foram essas reproduções. Em geral, faixas que estão sendo muito reproduzidas agora terão maior popularidade do que faixas que foram muito reproduzidas no passado. Faixas duplicadas (por exemplo, a mesma faixa de um single e de um álbum) são avaliadas independentemente. A popularidade do artista e do álbum é derivada matematicamente da popularidade das faixas.  
- **duration_ms**: O comprimento da faixa em milissegundos  
- **explicit**: Indica se a faixa possui letras explícitas (*true* = sim, possui; *false* = não possui ou é desconhecido)  
- **danceability**: Descreve o quão adequada a faixa é para dançar, com base em uma combinação de elementos musicais como tempo, estabilidade do ritmo, força da batida e regularidade geral. Um valor de 0.0 é menos dançante e 1.0 é mais dançante  
- **energy**: Medida de 0.0 a 1.0 que representa uma avaliação perceptiva de intensidade e atividade. Normalmente, faixas enérgicas parecem rápidas, altas e barulhentas. Por exemplo, *death metal* tem alta energia, enquanto um prelúdio de Bach pontua baixo nessa escala  
- **key**: A tonalidade em que a faixa está. Os números inteiros correspondem às alturas musicais usando a notação padrão de *Pitch Class*. Se nenhuma tonalidade foi detectada, o valor é -1  
- **loudness**: A intensidade geral de uma faixa em decibéis (dB)  
- **mode**: Indica a modalidade (maior ou menor) de uma faixa, o tipo de escala de onde vem seu conteúdo melódico. Maior é representado por 1 e menor é 0  
- **speechiness**: Detecta a presença de palavras faladas em uma faixa. Quanto mais a gravação for exclusivamente falada (por exemplo, programa de rádio, livro de áudio, poesia), mais próximo de 1.0 será o valor desse atributo. Valores acima de 0.66 descrevem faixas que provavelmente são inteiramente de palavras faladas. Valores entre 0.33 e 0.66 descrevem faixas que podem conter tanto música quanto fala, seja em seções ou sobrepostas, incluindo casos como música rap. Valores abaixo de 0.33 provavelmente representam música e outras faixas com pouca ou nenhuma fala  
- **acousticness**: Uma medida de confiança de 0.0 a 1.0 se a faixa é acústica. 1.0 representa alta confiança de que a faixa é acústica  
- **instrumentalness**: Prevê se uma faixa não contém vocais. Sons como “Ooh” e “aah” são tratados como instrumentais neste contexto. Faixas de rap ou faladas são claramente “vocais”. Quanto mais próximo de 1.0, maior a probabilidade de a faixa não conter conteúdo vocal  
- **liveness**: Detecta a presença de uma plateia na gravação. Valores mais altos de *liveness* representam uma maior probabilidade de que a faixa tenha sido gravada ao vivo. Um valor acima de 0.8 indica grande probabilidade de a faixa ser ao vivo  
- **valence**: Uma medida de 0.0 a 1.0 que descreve o grau de positividade musical transmitido por uma faixa. Faixas com alta *valência* soam mais positivas (ex.: felizes, alegres, eufóricas), enquanto faixas com baixa valência soam mais negativas (ex.: tristes, deprimidas, raivosas)  
- **tempo**: O tempo estimado geral de uma faixa em batidas por minuto (BPM). Em terminologia musical, *tempo* é a velocidade ou andamento de uma peça, derivada diretamente da duração média da batida  
- **time_signature**: Uma estimativa de compasso. O compasso (ou fórmula de compasso) especifica quantas batidas existem em cada compasso (ou medida). Varia de 3 a 7, indicando compassos de 3/4 até 7/4  
- **track_genre**: O gênero ao qual a faixa pertence
